## Работаем с картинками в Python

Библиотека [Pillow](https://pypi.org/project/Pillow/) это очень мощный, быстрый и удобный инструмент для обработки графики в python.

В сегодняшнем примере применим ее для конвертации картинки в черо белую, обратим цвета, вертикально отзеркалим и сохраним в файл.

Метод `Image.open('resources/img/cat.jpg')` считывает картинку из папки.

`img.convert('L')` трансформирует картинку в черно-белую.

`ImageOps.invert(image)` обращает цвета.

`img.transpose(Image.FLIP_LEFT_RIGHT)` отзеркаливает картинку.

`img.save('grayscale_cat.jpg')` сохранит картинку на диске.

Больше официальных примеров и документацию можно посмотреть [тут](https://pillow.readthedocs.io/en/stable/handbook/tutorial.html).

### Установка библиотеки
`pip install pillow`

```python
from PIL import Image, ImageOps

# Открыть файл картинки
img = Image.open('resources/img/cat.jpg')

# Конвертирование в черно-белую палитру
img = img.convert('L')

# Обратить цвета
img = ImageOps.invert(img)

# Отзеркалить по горизонтали
img = img.transpose(Image.FLIP_LEFT_RIGHT)

# Сохранить в файл
img.save('grayscale_cat.jpg')
```
